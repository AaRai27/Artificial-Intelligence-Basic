# Initialize Random Population

```python
import random as rand


def generate_kromosom(length,min,max):
# length = Banyak gen dalam 1 kromosom
        return [rand.randint(min,max) for x in range(length)]

# generate_kromosom(4,0,9)

def generate_population(count,length,min,max):
# count = banyak kromosom dalam populasi
# length = banyak gen dalam 1 kromosom
    return [generate_kromosom(length,min,max) for x in range(count)]
```

# Fitness Evaluation

```python
# Batasan :  (-1 <= x1 <= 2) and (-1 <= x2 <= 1)
# Rumus Decode Integer --> x = rmin + ((rmax - rmin)/9*(10^-1 + 10^-2 +...+10^-n/2)) * (gen1*)
def decode_kromosom(kromosom):
    x1 = -1 + (( 2 - (-1)) / 9*( 10**-1 + 10**-2 + 10**-3)) + ((kromosom[0] * 10**-1) + (kromosom[1] * 10**-2) + (kromosom[2] * 10**-3))
    x2 = -1 + (( 1 - (-1)) / 9*( 10**-1 + 10**-2 + 10**-3)) + ((kromosom[3] * 10**-1) + (kromosom[4] * 10**-2) + (kromosom[5] * 10**-3))
    return [x1,x2]

import math
def fitness(kromosom):
    x = decode_kromosom(kromosom)
    fitness_value = -(( math.cos(x[0]) ) * ( math.sin(x[1]) ) - ( x[0] / ((x[1]**2) + 1) ))
    return fitness_value

def fitness_all(kromosom,population_size):
    fit_krom = []
    for i in range(population_size):
        fit_krom.append(fitness(kromosom[i]))
    return fit_krom
```

# Selection

```python
def tournament_selection(population,tour_size, population_size):
    best = []
    for i in range(0, tour_size):
        krom = population[rand.randint(0,population_size-1)]
        if (best == [] or fitness(krom) > fitness(best)):
            best = krom
    return best
```

# Crossover

```python
def crossover(parent1,parent2,prob_c):
    r = rand.random()
    if(r < prob_c):
        switch_point = rand.randint(0,5)
        parent1[switch_point:],parent2[switch_point:] = parent2[switch_point:],parent1[switch_point:]
    return parent1,parent2
```

# Mutation

```python
def mutation(parent1,parent2,prob_m):
    r = rand.random()
    if( r < prob_m ):
        parent1[rand.randint(0,5)] = rand.randint(0,9)
        parent2[rand.randint(0,5)] = rand.randint(0,9)
    return parent1,parent2
```

# Variabel

```python
generasi = 100
kromosom_size = 6
population_size = 4
tour_size = 5
prob_c = 0.6
prob_m = 0.01
```

# Steady-State

```python
i = 0
population = generate_population(population_size,kromosom_size,0,9)
# print(population)
while i < generasi:
    print("Populasi Generasi " + str(i) + " = ")
    print(population)
    fit_all = fitness_all(population,population_size)
    print("Fitness All = " + str(fit_all))
    parent1 = tournament_selection(population,tour_size,population_size)
    print("Parent 1 = "+ str(parent1))
    parent2 = tournament_selection(population,tour_size,population_size)
    while(parent1 == parent2):
        parent2 = tournament_selection(population,tour_size,population_size)
    print("Parent 2 = "+ str(parent2))
    child = crossover(parent1,parent2,prob_c)
    print("crossover = "+ str(child))
    child = mutation(child[0],child[1],prob_m)
    print("Mutasi = "+ str(child))
    print("Populasi Sekarang = " + str(population))
    print()
    i+=1

print("=============== Hasil =============================")
    
best_kromosom = fit_all.index(max(fit_all))
print("Kromosom Terbaik = " + str(population[best_kromosom]))
print("Hasil Decode Kromosom = " + str(decode_kromosom(population[best_kromosom])))
print("Fitness Kromosom Terbaik = " + str(fit_all[best_kromosom]))
```
