# Membership Function Pengeluaran

```python
import matplotlib.pyplot as plt

# Function Pengeluaran Tinggi
pengeluaran_tinggi = 12
pengeluaran_tidak_tinggi = 3
def funcPengeluaranTinggi(pengeluaran):
    if(pengeluaran <= pengeluaran_tidak_tinggi):
        return 0
    elif(pengeluaran > pengeluaran_tinggi):
        return 1
    elif(pengeluaran > pengeluaran_tidak_tinggi and pengeluaran <= pengeluaran_tinggi):
        return (pengeluaran-pengeluaran_tidak_tinggi)/(pengeluaran_tinggi-pengeluaran_tidak_tinggi)

    
# Membership Function Pengeluaran Rendah
pengeluaran_rendah = 3
pengeluaran_tidak_rendah = 12
def funcPengeluaranRendah(pengeluaran):
    if(pengeluaran <= pengeluaran_rendah):
        return 1
    elif(pengeluaran > pengeluaran_tidak_rendah):
        return 0
    elif(pengeluaran > pengeluaran_rendah and pengeluaran <= pengeluaran_tidak_rendah):
        return (pengeluaran_tidak_rendah-pengeluaran)/(pengeluaran_tidak_rendah-pengeluaran_rendah)
    
    
# Membership Function Pengeluaran Lumayan
pengeluaran_lumayan_min = 6
pengeluaran_lumayan_max = 12
pengeluaran_besar = 9
pengeluaran_kecil = 3
def funcPengeluaranLumayan(pengeluaran):
    if(pengeluaran <= pengeluaran_kecil or pengeluaran > pengeluaran_besar):
        return 0
    elif(pengeluaran > pengeluaran_kecil and pengeluaran <= pengeluaran_lumayan_min):
        return (pengeluaran-pengeluaran_kecil)/(pengeluaran_lumayan_min-pengeluaran_kecil)
    elif (pengeluaran > pengeluaran_lumayan_min and pengeluaran <= pengeluaran_lumayan_max):
        return 1
    elif (pengeluaran > pengeluaran_lumayan_max and pengeluaran <= pengeluaran_besar):
        return (pengeluaran_besar-pengeluaran)/(pengeluaran_besar-pengeluaran_lumayan_max)

# ---------------- Uji Coba -------------------------
# Pengeluaran Tinggi
x1 = [0,3,6,9,12,15]
y1 = [funcPengeluaranTinggi(0),funcPengeluaranTinggi(3),funcPengeluaranTinggi(6),funcPengeluaranTinggi(9),funcPengeluaranTinggi(12),funcPengeluaranTinggi(15)]
# Pengeluaran Lumayan
x2 = [1,3,6,9,12,14]
y2 = [funcPengeluaranLumayan(1),funcPengeluaranLumayan(3),funcPengeluaranLumayan(6),funcPengeluaranLumayan(9),funcPengeluaranLumayan(12),funcPengeluaranLumayan(14)]
# Pengeluaran Rendah
x3 = [0,3,6,9,12,15]
y3 = [funcPengeluaranRendah(0),funcPengeluaranRendah(3),funcPengeluaranRendah(6),funcPengeluaranRendah(9),funcPengeluaranRendah(12),funcPengeluaranRendah(15)]
plt.plot(x1,y1, label='Pengeluaran Tinggi') 
plt.plot(x2,y2, label='Pengeluaran Lumayan')
plt.plot(x3,y3, label='Pengeluaran Rendah')
plt.ylabel('Nilai Fuzzy')
plt.xlabel('Pengeluaran')
plt.legend()
```

# Membership Function Gaji

```python
import matplotlib.pyplot as plt


# Gaji Dalam Satuan Juta

# Membership Function Gaji Besar
gaji_besar = 12
gaji_tidak_besar = 3
def funcGajiBesar(gaji):
    # Range gaji 0-5
    if(gaji <= gaji_tidak_besar):
        return 0
    # Range gaji > 10
    elif(gaji > gaji_besar):
        return 1
    # Range gaji 5.1 - 10
    elif(gaji > gaji_tidak_besar and gaji <= gaji_besar):
        return (gaji-gaji_tidak_besar)/(gaji_besar-gaji_tidak_besar)

# Membership Function Gaji Kecil
gaji_kecil = 3
gaji_tidak_kecil = 12
def funcGajiKecil(gaji):
    # Range Gaji 0 - 6
    if(gaji <= gaji_kecil):
        return 1
    # Range Gaji > 9
    elif(gaji > gaji_tidak_kecil):
        return 0
    # Range Gaji 6.1 - 9
    elif(gaji > gaji_kecil and gaji <= gaji_tidak_kecil):
        return (gaji_tidak_kecil-gaji)/(gaji_tidak_kecil-gaji_kecil)

    
    
# Membership Function Gaji Cukup
gaji_cukup_min = 6
gaji_cukup_max = 9
gaji_tinggi = 12
gaji_rendah = 3
def funcGajiCukup(gaji):
    # Gaji < 6.5 juta atau gaji > 9.5
    if(gaji <= gaji_rendah or gaji > gaji_tinggi):
        return 0
    # Range Gaji 6.1 - 
    elif(gaji > gaji_rendah and gaji <= gaji_cukup_min):
        return (gaji-gaji_rendah)/(gaji_cukup_min-gaji_rendah)
    elif (gaji > gaji_cukup_min and gaji <= gaji_cukup_max):
        return 1
    elif (gaji > gaji_cukup_max and gaji <= gaji_tinggi):
        return (gaji_tinggi-gaji)/(gaji_tinggi-gaji_cukup_max)

    
# ------------------- Uji Coba ------------------------
# Gaji Besar
a1 = [0,3,6,9,12,15]
b1 = [funcGajiBesar(0),funcGajiBesar(3),funcGajiBesar(6),funcGajiBesar(9),funcGajiBesar(12),funcGajiBesar(15)]
# Gaji Kecil
a2 = [0,3,6,9,12,15]
b2 = [funcGajiKecil(0),funcGajiKecil(3),funcGajiKecil(6),funcGajiKecil(9),funcGajiKecil(12),funcGajiKecil(15)]
# Gaji Cukup
a3 = [1,3,6,9,12,14]
b3 = [funcGajiCukup(1),funcGajiCukup(3),funcGajiCukup(7),funcGajiCukup(9),funcGajiCukup(12),funcGajiCukup(14)]

plt.plot(a1,b1, label='Gaji Besar') 
plt.plot(a2,b2, label='Gaji Kecil')
plt.plot(a3,b3, label='Gaji Cukup')
plt.ylabel('Nilai Fuzzy')
plt.xlabel('Gaji')
plt.legend()
```

# Fuzzy Rules
### PENGELUARAN - GAJI - SCORE
* Tinggi - Besar - Dipertimbangkan
* Tinggi - Cukup - Diterima
* Tinggi - Kecil - Diterima
* Lumayan- Besar - Dipertimbangkan  
* Lumayan- Cukup - Dipertimbangkan
* Lumayan- Kecil - Diterima
* Rendah - Besar - Ditolak
* Rendah - Cukup - Ditolak
* Rendah - Kecil - Dipertimbangkan


```python
# ling_pengeluaran = nilai linguistik pengeluaran
# ling_gaji = nilai linguistik gaji
def fuzzyRules(ling_pengeluaran, ling_gaji, pengeluaran, gaji, id):
    fuz_output = []
    if(ling_pengeluaran[0] == 'tinggi' and ling_gaji[0] == 'besar'):
        fuz_output.append(['dipertimbangkan',min(pengeluaran[id][1],gaji[id][1])])
    if(ling_pengeluaran[0] == 'tinggi' and ling_gaji[1] == 'cukup'):
        fuz_output.append(['diterima',min(pengeluaran[id][1],gaji[id][2])])
    if(ling_pengeluaran[0] == 'tinggi' and ling_gaji[2] == 'kecil'):
        fuz_output.append(['diterima',min(pengeluaran[id][1],gaji[id][3])])
    if(ling_pengeluaran[1] == 'lumayan' and ling_gaji[0] == 'besar'):
        fuz_output.append(['dipertimbangkan',min(pengeluaran[id][2],gaji[id][1])])
    if(ling_pengeluaran[1] == 'lumayan' and ling_gaji[1] == 'cukup'):
        fuz_output.append(['dipertimbangkan',min(pengeluaran[id][2],gaji[id][2])])
    if(ling_pengeluaran[1] == 'lumayan' and ling_gaji[2] == 'kecil'):
        fuz_output.append(['diterima',min(pengeluaran[id][2],gaji[id][3])])
    if(ling_pengeluaran[2] == 'rendah' and ling_gaji[0] == 'besar'):
        fuz_output.append(['ditolak',min(pengeluaran[id][3],gaji[id][1])])
    if(ling_pengeluaran[2] == 'rendah' and ling_gaji[1] == 'cukup'):
        fuz_output.append(['ditolak',min(pengeluaran[id][3],gaji[id][2])])
    if(ling_pengeluaran[2] == 'rendah' and ling_gaji[2] == 'kecil'):
        fuz_output.append(['dipertimbangkan',min(pengeluaran[id][3],gaji[id][3])])
    
    return fuz_output
```

# Defuzzifikasi (Sugeno)

```python
# Konstanta Diterima = 90, Dipertimbangkan = 70, Ditolak = 50
def defuzzifikasi(inferensi):
    return ((inferensi[0]*90) + (inferensi[1]*70) + (inferensi[2]*50))/(inferensi[0]+inferensi[1]+inferensi[2])

plt.axvline(x=90, color = 'green', label='Diterima')
plt.axvline(x=70, color='blue', label='Dipertimbangkan')
plt.axvline(x=50, color='red', label='Ditolak')
plt.legend()
```

```python
linguistik_pengeluaran = ['tinggi','lumayan','rendah']
linguistik_gaji = ['besar','cukup','kecil']


nilai_pengeluaran = []
for i in range(len(df)):
    value = []
    value.append(df["Id"][i])
    value.append(funcPengeluaranTinggi(df["Pengeluaran"][i]))
    value.append(funcPengeluaranLumayan(df["Pengeluaran"][i]))
    value.append(funcPengeluaranRendah(df["Pengeluaran"][i]))
    nilai_pengeluaran.append(value)
    
nilai_gaji = []
for i in range(len(df)):
    value = []
    value.append(df["Id"][i])
    value.append(funcGajiBesar(df["Penghasilan"][i]))
    value.append(funcGajiCukup(df["Penghasilan"][i]))
    value.append(funcGajiKecil(df["Penghasilan"][i]))
    nilai_gaji.append(value)
    
defuz = []
for i in range(len(df)):
    fuzzy = fuzzyRules(linguistik_pengeluaran, linguistik_gaji, nilai_pengeluaran, nilai_gaji, i)
    inferensi = []
    terima = []
    timbang = []
    tolak = []

    for j in range(len(fuzzy)):
        if(fuzzy[j][0] == 'diterima'):
            terima.append(fuzzy[j][1])
        if(fuzzy[j][0] == 'dipertimbangkan'):
            timbang.append(fuzzy[j][1])
        if(fuzzy[j][0] == 'ditolak'):
            tolak.append(fuzzy[j][1])
    
    inferensi.append(max(terima))
    inferensi.append(max(timbang))
    inferensi.append(max(tolak))
    
    defuz.append([defuzzifikasi(inferensi),i+1]) # i adalah mahasiswa. ditambah 1 karena agar index dimulai dari 1

defuz.sort(reverse=True)
# print(defuz[1][1])
    
    
```
